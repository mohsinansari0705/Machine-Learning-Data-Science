Importing all essential Libraries to plot all the kinds of Distributions/plots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson
```

## 1. Implementation of Normal Distribution in Python

```python
# Data used to plot Normal Distribution
mean = 10 # mean
std = 2 # standard deviation
size = 1000 # size of the distribution

data = np.random.normal(loc=mean, scale=std, size=size)

sns.histplot(data, kde=True, stat='density', bins=30, color='skyblue', linewidth=0.8)

plt.title(f"Normal Distribution (μ={mean}, σ={std})")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```

## 2. Implementation of Binomial Distribution in Python

```python
n = 10 # No. of trials
p = 0.5 # Probability of getting success
x = np.arange(0, n+1) # values from 0 to 10

# Probability Mass Function (PMF) of getting 0 to 10 success
pmf = binom.pmf(x, n, p)

# Cumulative Distribution Function (CDF) of getting at most n successes (0 to 10)
cdf = binom.cdf(x, n, p)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(8, 10))

# Binomial Distribution PMF
ax0.bar(x, pmf, color='skyblue', edgecolor='black')
ax0.set(title="Binomial Distribution PMF (n=10, p=0.5)",
        xlabel="Number of Successes (x)",
        ylabel="Probability")
ax0.grid(axis='y', linestyle='--', alpha=0.7)

# Binomial Distribution CDF
ax1.plot(x, cdf, color='purple', marker='o', linestyle='-', linewidth=2)
ax1.set(title="Binomial Distribution CDF (n=10, p=0.5)",
        xlabel="Number of Successes (x)",
        ylabel="Cumulative Probability")
ax1.grid(True)

plt.show()
```

## 3. Implementation of Normal Distribution in Python

```python
a, b = 0, 25 # Interval [0, 25]

samples = np.random.uniform(a, b, 1000)
x = np.linspace(a, b, 1000)

plt.hist(samples, bins=30, density=True, color='lightgreen', alpha=0.7)
plt.plot(x, np.ones_like(x)/(b-a), 'r-', lw=2)

plt.title("Uniform Distribution PDF")
plt.xlabel("X")
plt.ylabel("Probability Density")

plt.show()
```

## 4. Implementation of Exponential Distribution in Python

```python
# set the rate parameter lambda
lam = 1.5

# generate Exponential data
data = np.random.exponential(1/lam, 1000)

# plot the KDE and histogram
sns.histplot(data, kde=True, stat='density', color='skyblue', linewidth=0.7)

plt.title("Exponential Distribution (lambda=1.5)")
plt.xlabel("Time")
plt.ylabel("Density")

plt.show()
```

## 5. Implementation of Poisson Distribution in Python

```python
lambda_val = 3
k = np.arange(0, 10)

pmf = poisson.pmf(k, lambda_val)
cdf = poisson.cdf(k, lambda_val)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(8, 10))

ax0.bar(k, pmf, color='lightgreen', edgecolor='black')
ax0.set(title="Poisson Distribution PMF (λ=3)",
        xlabel="Number of Events (k)",
        ylabel="Probability")
ax0.grid(axis='y', linestyle='--', alpha=0.7)

ax1.plot(k, cdf, color='purple', marker='o', linestyle='-', linewidth=2)
ax1.set(title="Poisson Distribution CDF (λ=3)",
        xlabel="Number of Events (k)",
        ylabel="Cumulative Probability")
ax1.grid(True)

plt.show()
```
